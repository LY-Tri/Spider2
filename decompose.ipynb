```python
import json
import openai

# load the json file
with open("./df_sql.json", "r") as f:
    df_sql = json.load(f)
```

```python
from decompose import SQLDecomposer

subqueries_count = []

for d in df_sql:
    sql = d["sql"]
    decomposer = SQLDecomposer(sql, dialect="snowflake")
    subqueries = decomposer.queries
    subqueries_count.append(len(subqueries))

# print the mean, max, min, and median of the subqueries_count
print(f"Mean: {sum(subqueries_count) / len(subqueries_count)}")
print(f"Max: {max(subqueries_count)}")
print(f"Min: {min(subqueries_count)}")
print(f"Median: {sorted(subqueries_count)[len(subqueries_count) // 2]}")
```

```python
# query openai model to decompose the sql
server = openai.OpenAI(
    api_key=open("./openai.key").read()
)

# query the model
response = server.chat.completions.create(
    model="gpt-5-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant that decomposes a complex SQL query into a sequence of executable SQL subqueries."},
        {"role": "user", "content": sql}
    ]
)

# print the response
print(response.choices[0].message.content)
```
